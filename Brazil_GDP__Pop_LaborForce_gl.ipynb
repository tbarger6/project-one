## GDP Per Capita vs. Population ages 15-64 and Labor Force, Total - Brazil

```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
import numpy as np
```

```python

#Find base url for api
base_url = "http://api.worldbank.org/v2"
base_url
```

```python
#Pick indicator you want to search for 
#GDP per Capita (current US$) = NY.GDP.PCAP.CD  


indicator = "NY.GDP.PCAP.CD"
```

```python
#Pick a country
country = "BR"
```

```python
#Choose desired format
response_format = "json"
```

```python
request_url = f"{base_url}/country/{country}/indicator/{indicator}?format={response_format}"
request_url
```

```python
response = requests.get(request_url).json()
```

```python
response
```

```python
final_response = response[1]
final_response
```

```python
df = pd.DataFrame(final_response)
df
```

```python
brazil_gdp = df[["date","value"]]
brazil_gdp = brazil_gdp.sort_values("date", ascending=True)
brazil_gdp.head()
```

```python
plt.plot(brazil_gdp["date"],brazil_gdp["value"])
```

##Make API call for Population

```python
#Pick indicator you want to search for 
#Population ages 15-64 (total population) = SP.POP.1564.TO

indicator = "SP.POP.1564.TO"
```

```python
request_url = f"{base_url}/country/{country}/indicator/{indicator}?format={response_format}"
request_url
```

```python

response2 = requests.get(request_url).json()
```

```python
second_response = response2[1]
second_response
```

```python
df = pd.DataFrame(second_response)
df
```

```python
brazil_pop = df[["date","value"]]
brazil_pop = brazil_pop.sort_values("date", ascending=True)
brazil_pop.head()
```

```python
plt.plot(brazil_pop["date"],brazil_pop["value"])
```

```python
##Make call for labor Force Total
```

```python
#Pick indicator you want to search for 
#Labor force, total = SL.TLF.TOTL.IN

indicator = "SL.TLF.TOTL.IN"
```

```python
request_url = f"{base_url}/country/{country}/indicator/{indicator}?format={response_format}"
request_url
```

```python
response3 = requests.get(request_url).json()
```

```python
third_response = response3[1]
third_response
```

```python
df = pd.DataFrame(third_response)
df
```

```python
brazil_labor= df[["date","value"]]
brazil_labor = brazil_labor.sort_values("date", ascending=True)
brazil_labor.head()
```

```python
plt.plot(brazil_labor["date"],brazil_labor["value"])
```

```python
first_merge = pd.merge(brazil_gdp, brazil_pop, on="date", how="inner")
first_merge
```

```python
merge_table = pd.merge(first_merge, brazil_labor, on="date", how="inner")
merge_table
```

```python
# Clean the data
merge_table = merge_table.rename(columns = {"date": "Date", "value_x": "GDP Per Capita $", "value_y": "Total Population", "value": "Total Labor Force"})
merge_table
```

```python
merge_table = merge_table.dropna()
merge_table
```

```python
#Verify the values are numeric
merge_table.dtypes
```

```python
##SCATTER PLOT To Compare GDP Per Capita and Total Population of Brazil between the ages 15 - 64
```

```python
#import module to run linregress
import scipy.stats as st
from scipy.stats import linregress
```

```python
# Add the linear regression equation and line to scatter plot
x_values = merge_table['Total Population']
y_values = merge_table['GDP Per Capita $']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.xticks(rotation=45)
plt.plot(x_values,regress_values,"r-")
plt.title('GDP Per Capita (current Brazil$) vs. Total Population (ages 15-64)')
plt.xlabel('Total Population (ages 15-64)')
plt.ylabel('GDP Per Capita (current Brazil$)')
#Turn off the scientific notation (ie8)
plt.ticklabel_format(style="Plain")
# Calculate the correlation coefficient between the two variables
print(f"The correlation coefficient for this relationship is {round(st.pearsonr(x_values,y_values)[0],2)}")
plt.savefig("GDPvsTotalPopulation.png")
plt.show()
```
