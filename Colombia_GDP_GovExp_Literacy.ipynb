```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
import numpy as np
```

```python
#Find base url for api
base_url = "http://api.worldbank.org/v2"
base_url
```

```python
#Pick indicator you want to search for 
#GDP per Capita (current US$) = NY.GDP.PCAP.CD  


indicator = "NY.GDP.PCAP.CD"
```

```python
#Pick a country
country = "CO"
```

```python
#Choose desired format
response_format = "json"
```

```python
request_url = f"{base_url}/country/{country}/indicator/{indicator}?format={response_format}"
request_url
```

```python
response = requests.get(request_url).json()
```

```python
response
```

```python
final_response = response[1]
final_response
```

```python
df = pd.DataFrame(final_response)
df
```

```python
colombia_gdp = df[["date","value"]]
colombia_gdp = colombia_gdp.sort_values("date", ascending=True)
colombia_gdp.head()
```

```python
plt.plot(colombia_gdp["date"],colombia_gdp["value"])
```

```python
#Pick indicator you want to search for 
#government expenditure per student, primary (% of GDP per capita) ID: SE.XPD.PRIM.PC.ZS

indicator = "SE.XPD.PRIM.PC.ZS"
```

```python
request_url = f"{base_url}/country/{country}/indicator/{indicator}?format={response_format}"
request_url
```

```python
response2 = requests.get(request_url).json()
```

```python
second_response = response2[1]
second_response
```

```python
df = pd.DataFrame(second_response)
df
```

```python
colombia_govexp = df[["date","value"]]
colombia_govexp = colombia_govexp.sort_values("date", ascending=True)
colombia_govexp.head()
```

```python
plt.plot(colombia_govexp["date"],colombia_govexp["value"])
```

```python
##Make call for Literacy Rate, adult total (% of people ages 15 and above)
#Pick indicator you want to search for 
#literacy rate, adult total (% of people ages 15 and above) ID: SE.ADT.LITR.ZS

indicator = "SE.ADT.LITR.ZS"
```

```python
request_url = f"{base_url}/country/{country}/indicator/{indicator}?format={response_format}"
request_url
```

```python
response3 = requests.get(request_url).json()
```

```python
third_response = response3[1]
third_response
```

```python
df = pd.DataFrame(third_response)
df
```

```python
colombia_literacy= df[["date","value"]]
colombia_literacy = colombia_literacy.sort_values("date", ascending=True)
colombia_literacy.head()
```

```python
plt.plot(colombia_literacy["date"],colombia_literacy["value"])
```

```python
first_merge = pd.merge(colombia_gdp, colombia_govexp, on="date", how="inner")
first_merge
```

```python
merge_table = pd.merge(first_merge, colombia_literacy, on="date", how="inner")
merge_table
```

```python
# Clean the data
merge_table = merge_table.rename(columns = {"date": "Date", "value_x": "GDP Per Capita $", "value_y": "Gov Expenditure", "value": "Literacy Rate"})
merge_table
```

```python
merge_table = merge_table.dropna()
merge_table
```

```python
#Verify the values are numeric
merge_table.dtypes
```

```python
##SCATTER PLOT To Compare GDP Per Capita and Gov Expenditure per student
#import module to run linregress
import scipy.stats as st
from scipy.stats import linregress
```

```python

# Add the linear regression equation and line to scatter plot
x_values = merge_table['Gov Expenditure']
y_values = merge_table['GDP Per Capita $']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)

plt.plot(x_values,regress_values,"r-")
plt.title('GDP Per Capita vs. Goverment Expenditure per Student')
plt.xlabel('Government Expenditure per Student')
plt.ylabel('GDP Per Capita (current Colombia$)')
#Turn off the scientific notation (ie8)
plt.ticklabel_format(style="Plain")
# Calculate the correlation coefficient between the two variables
print(f"The correlation coefficient for this relationship is {round(st.pearsonr(x_values,y_values)[0],2)}")
plt.savefig("GDPvsGovExpPerStudent.png")
plt.show()
```
