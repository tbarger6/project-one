## GDP Growth (annual%) vs. Labor Force, Total - Brazil

```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
import numpy as np
```

```python

#Find base url for api
base_url = "http://api.worldbank.org/v2"
base_url
```

```python
#Pick indicator you want to search for  GDP growth(annual %) = NY.GDP.MKTP.KD.ZG, 

indicator = "NY.GDP.MKTP.KD.ZG"
```

```python
#Pick a country
country = "BR"
```

```python
#Choose desired format
response_format = "json"
```

```python
request_url = f"{base_url}/country/{country}/indicator/{indicator}?format={response_format}"
request_url
```

```python
response = requests.get(request_url).json()
```

```python
response
```

```python
final_response = response[1]
final_response
```

```python
df = pd.DataFrame(final_response)
df
```

```python
brazil_gdp = df[["date","value"]]
brazil_gdp = brazil_gdp.sort_values("date", ascending=True)
brazil_gdp.head()
```

```python
plt.plot(brazil_gdp["date"],brazil_gdp["value"])
```
