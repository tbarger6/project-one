```python
## Analysis on GDP Per Capita vs Education, Population ages 0-14 (% of total population) and 
#School Enrollment, Primary, Female (% gross) - Tanzania
```

```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
import numpy as np
```

```python

#Find base url for api
base_url = "http://api.worldbank.org/v2"
base_url
```

```python
#Pick indicator you want to search for 
#GDP per Capita (current US$) = NY.GDP.PCAP.CD  


indicator = "NY.GDP.PCAP.CD"
```

```python
#Pick a country
country = "TZ"
```

```python
#Choose desired format
response_format = "json"
```

```python
request_url = f"{base_url}/country/{country}/indicator/{indicator}?format={response_format}"
request_url
```

```python
response = requests.get(request_url).json()
```

```python
response
```

```python
final_response = response[1]
final_response
```

```python

df = pd.DataFrame(final_response)
df
```

```python
tanzania_gdp = df[["date","value"]]
tanzania_gdp = tanzania_gdp.sort_values("date", ascending=True)
tanzania_gdp.head()
```

```python
plt.plot(tanzania_gdp["date"],tanzania_gdp["value"])
```

```python
#Pick indicator you want to search for 
#Population ages 0-14 (% of total population) = SP.POP.0014.TO.ZS 
indicator = "SP.POP.0014.TO.ZS"
```

```python
request_url = f"{base_url}/country/{country}/indicator/{indicator}?format={response_format}"
request_url
```

```python
response2 = requests.get(request_url).json()
```

```python
second_response = response2[1]
second_response
```

```python
df = pd.DataFrame(second_response)
df
```

```python
tanzania_pop = df[["date","value"]]
tanzania_pop = tanzania_pop.sort_values("date", ascending=True)
tanzania_pop.head()
```

```python
plt.plot(tanzania_pop["date"],tanzania_pop["value"])
```

```python
#Pick indicator you want to search for 
#primary enrollment on education female, total = SE.PRM.ENRR.FE 
indicator = "SE.PRM.ENRR.FE"
```

```python
request_url = f"{base_url}/country/{country}/indicator/{indicator}?format={response_format}"
request_url
```

```python
response3 = requests.get(request_url).json()
```

```python
third_response = response3[1]
third_response
```

```python
df = pd.DataFrame(third_response)
df
```

```python
tanzania_enroll = df[["date","value"]]
tanzania_enroll = tanzania_enroll.sort_values("date", ascending=True)
tanzania_enroll.head()
```

```python
plt.plot(tanzania_enroll["date"],tanzania_enroll["value"])
```

```python
first_merge = pd.merge(tanzania_gdp, tanzania_pop, on='date', how='inner')
first_merge
```

```python
merge_table = pd.merge(first_merge, tanzania_enroll, on='date', how='inner')
merge_table
```

```python
merge_table = merge_table.rename(columns = {"data": "Date", "value_x": "GDP", 
                                            "value_y": "Total Population(0-14)", "value": "Enrollment"})
merge_table                                         
```

```python
merge_table = merge_table.dropna()
merge_table
```

```python
#Verify the values are numeric
merge_table.dtypes
```

```python
#Scatter Plot To Compare GDP Per Capita and population education of Tanzania between the ages 0-14
```

```python
#import module to run linregress
import scipy.stats as st
from scipy.stats import linregress
```

```python
# Add the linear regression equation and line to scatter plot
x_values = merge_table['Total Population(0-14)']
y_values = merge_table['GDP']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.xticks(rotation=45)
plt.plot(x_values,regress_values,"r-")
plt.title('GDP Per Capita (current Tanzania) vs. Total Population(0-14)')
plt.xlabel('Total Population(0-14)')
plt.ylabel('GDP Per Capita (current Tanzania)')
#Turn off the scientific notation (ie8)
plt.ticklabel_format(style="Plain")
# Calculate the correlation coefficient between the two variables
print(f"The correlation coefficient for this relationship is {round(st.pearsonr(x_values,y_values)[0],2)}")
plt.savefig("GDPvsTotalPopulation(0-14).png")
plt.show()
```

```python
#Scatter Plot To Compare GDP Per Capita and school enrollment primary, female
# Add the linear regression equation and line to scatter plot
x_values = merge_table['Enrollment']
y_values = merge_table['GDP']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.xticks(rotation=45)
plt.plot(x_values,regress_values,"r-")
plt.title('GDP Per Capita (current Tanzania) vs. Primary Enrollment')
plt.xlabel('Enrollment')
plt.ylabel('GDP Per Capita (current Tanzania)')
#Turn off the scientific notation (ie8)
plt.ticklabel_format(style="Plain")
# Calculate the correlation coefficient between the two variables
print(f"The correlation coefficient for this relationship is {round(st.pearsonr(x_values,y_values)[0],2)}")
plt.savefig("GDPvsEnrollment.png")
plt.show()
```
