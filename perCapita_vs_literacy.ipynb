```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
import numpy as np
```

```python
#Find base url for api
base_url = "http://api.worldbank.org/v2"
base_url
```

```python

#Pick indicator you want to search for 
#GDP per Capita (current US$) = NY.GDP.PCAP.CD  


indicator = "NY.GDP.PCAP.CD"
```

```python
#Pick a country
country = "CO"
```

```python
#Choose desired format
response_format = "json"
```

```python
request_url = f"{base_url}/country/{country}/indicator/{indicator}?format={response_format}"
request_url
```

```python
response = requests.get(request_url).json()
```

```python
response
```

```python
final_response = response[1]
final_response
```

```python
df = pd.DataFrame(final_response)
df
```

```python
colombia_gdp = df[["date","value"]]
colombia_gdp = colombia_gdp.sort_values("date", ascending=True)
colombia_gdp.head()
```

```python
plt.plot(colombia_gdp["date"],colombia_gdp["value"])
```

```python
#Pick indicator you want to search for 
#literacy rate, adult total (% of people ages 15 and above) ID: SE.ADT.LITR.ZS

indicator = "SE.ADT.LITR.ZS"
```
